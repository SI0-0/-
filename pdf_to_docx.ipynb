```python
import os
from tkinter import Tk, filedialog
from PyPDF2 import PdfReader
from docx import Document

def pdf_to_docx_simple(pdf_path, docx_path):
    reader = PdfReader(pdf_path)
    doc = Document()

    for page in reader.pages:
        text = page.extract_text()
        if text:
            for line in text.splitlines():
                doc.add_paragraph(line)
            doc.add_page_break()

    doc.save(docx_path)

def select_pdf_and_convert():
    root = Tk()
    root.withdraw()

    pdf_path = filedialog.askopenfilename(
        title="PDF 파일을 선택하세요",
        filetypes=[("PDF files", "*.pdf")]
    )

    if not pdf_path:
        print("아무 파일도 선택하지 않았습니다.")
        return

    folder = os.path.dirname(pdf_path)
    filename = os.path.splitext(os.path.basename(pdf_path))[0]
    docx_path = os.path.join(folder, f"{filename}.docx")

    print("변환 중입니다...")
    pdf_to_docx_simple(pdf_path, docx_path)
    print(f"완료! {docx_path}")

select_pdf_and_convert()
```
